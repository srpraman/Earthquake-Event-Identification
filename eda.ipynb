```python
import numpy as np 
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
file_list=os.listdir("../data/waveforms/chunk2_ir_removed")
df=pd.read_csv("../data/csv_files/updated_chunk2.csv")
def magnitude(file_name,df):
    sl_no = int(file_name.split("_")[-1].split(".")[0])
    #return sl_no
    return df.iloc[sl_no]["source_magnitude"]
mag = []
for file in file_list:
    try:
        mag.append((magnitude(file,df)))
    except:
        #print(f"mag not available for {file}")
        pass
sns.set()
sns.histplot(data=mag,stat="count", binwidth=1)
plt.xlabel("Earthquake Magnitude")
plt.ylabel("Number of Occurrences")
plt.show()
plt.savefig("/home/sairaman/Desktop/poster/eq_count.png")
```
